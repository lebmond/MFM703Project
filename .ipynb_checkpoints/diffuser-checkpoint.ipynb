```python
import numpy as np
import torch
import matplotlib.pyplot as plt
```

```python
# Function to simulate stock price using Geometric Brownian Motion
def stock_price_simulation(initial_price, drift, volatility, time_horizon, num_steps):
    dt = time_horizon / num_steps
    time_vector = np.linspace(0, time_horizon, num_steps + 1)
    price = np.zeros(num_steps + 1)
    price[0] = initial_price

    for i in range(1, num_steps + 1):
        price[i] = price[i - 1] * np.exp((drift - 0.5 * volatility ** 2) * dt + volatility * np.sqrt(dt) * np.random.normal())

    return time_vector, price

# Parameters
initial_price = 150.0  # Initial stock price
drift = 0.2  # Drift of the stock (can be considered as expected return)
volatility = 0.3  # Volatility of the stock
time_horizon = 1.0  # Time horizon (in years)
num_steps = 252  # Number of time steps (trading days in a year)

# Simulate stock price
time, stock_price = stock_price_simulation(initial_price, drift, volatility, time_horizon, num_steps)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(time, stock_price, label='AAPL Stock Price Simulation')
plt.title('AAPL Stock Price Simulation using Geometric Brownian Motion')
plt.xlabel('Time (in years)')
plt.ylabel('Stock Price')
plt.legend()
plt.grid(True)
plt.show()
```
